```python
data = pd.read_excel('BCRD/imae.xlsx', skiprows = 9, skipfooter = 1, header = None)
data
```

```python
data = data[~(data[0].str.contains('Promedio') | data[0].isnull())]
```

```python
start_year = 2007

n_years = data.shape[0] // 12
years = np.array(list(range(start_year, start_year + n_years)))
years = np.repeat(years, [12], axis = 0)
last_year = np.repeat(2007 + n_years, month_last_year, axis = 0)
years = np.concatenate([years, last_year])

month_last_year = data.shape[0] % 12 # calculates how many months are available for last year
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'Octuber', 'November', 'December']
months = np.tile(months, [n_years + 1])[:data.shape[0]]
```

```python
col_idx_arr = list(zip(['serie_original', 'serie_original', 'serie_original', 'serie_original', 'serie_desestacionalizada', 'serie_desestacionalizada', 'serie_desestacionalizada', 'serie_desestacionalizada', 'serie_desestacionalizada', 'serie_tendencia_ciclo', 'serie_tendencia_ciclo', 'serie_tendencia_ciclo', 'serie_tendencia_ciclo', 'serie_tendencia_ciclo'],
                  ['indice', 'interanual', 'acumulada', 'prom_12_meses', 'indice', 'versus_periodo_anterior', 'interanual', 'acumulada', 'prom_12_meses', 'indice', 'versus_periodo_anterior', 'interanual', 'acumulada', 'prom_12_meses']))
col_idx = pd.MultiIndex.from_tuples(col_idx_arr)
```

```python
row_idx_arr = list(zip(years, months))
row_idx = pd.MultiIndex.from_tuples(row_idx_arr)
row_idx
```

```python
data.index = row_idx
data = data.loc[:, 2:]
data.columns = col_idx
```
